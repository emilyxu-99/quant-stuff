```python
# Emily and Ed's portfolio

import numpy as np
import pandas as pd
import IPython.display 
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
# Collecting data from the start of Black Thursday, the 2020 COVID-19 market crash.

%matplotlib inline

assets = ['NIO', '^GSPC', 'VGT', 'XLK', 'IWL', 'IVV', 'GSLC', 'BLK', 'VOO', 'SPY', 'SPGM', 'MSFT', 'XLG', 'QQQE', 'IVV',
         'NVDA', 'NDAQ', 'BABA', 'SPYG', 'GOOGL', 'TAN', 'QCLN', 'PBW', 'TSLA', 'HYLN', 'FSR', 'XPEV', 'LI', 'NIU', 'BYDDY',
         'ENPH', 'CSIQ', 'FSLR', 'PLUG', 'NOVA', 'SPWR', 'JD', 'PLTR', 'U', 'ZM', 'ZG', 'SNOW', 'PYPL', 'SQ',
         'FVRR', 'PINS', 'SHOP', 'CRWD', 'NET', 'NOW', 'TTD', 'PTON', 'ETSY', 'CRSP', 'AMD', 'TSM', 'CRSR', 'LOGI']

pf_data = pd.DataFrame()
for a in assets:
    pf_data[a] = wb.DataReader(a, data_source = 'yahoo', start = '2020-3-12')['Adj Close']
```

```python
# Normalize data such that each stock price starts at $100. NIO is clearly the superstar here.

(pf_data / pf_data.iloc[0] * 100).plot(figsize = (20, 20))
```

```python
num_assets = len(assets) - 1
num_assets
```

```python
weights = np.random.random(num_assets)
weights /= np.sum(weights)
weights

# Obtain num_assets weights that sum to 1
```

```python
# Expected portfolio return
np.sum(weights * returns.mean()) * 250
```

```python
# Variance
np.dot(weights.T, np.dot(returns.cov() * 250, weights))
```

```python
# Volatility
np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 250, weights)))
```

```python
pfolio_returns = []
pfolio_volatilities = []
weights = []

for x in range (1000):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    p_weights.append(weights)
    pfolio_returns.append(np.sum(weights * returns.mean()) * 250)
    pfolio_volatilities.append(np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 250, weights))))

pfolio_returns = np.array(pfolio_returns)
pfolio_volatilities = np.array(pfolio_volatilities)
```

```python
portfolios = pd.DataFrame({'Return': pfolio_returns, 'Volatility': pfolio_volatilities})
portfolios
```

```python
# Collect weights from monte carlo simulation and append to the corresponding ticker symbol

for counter, symbol in enumerate(pf_data.columns.tolist()):
    portfolios[symbol + ' weight'] = [w[counter] for w in p_weights]
```
